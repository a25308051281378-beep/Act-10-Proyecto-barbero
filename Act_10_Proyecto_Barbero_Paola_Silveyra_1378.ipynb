```python
# ==========================================
# PROYECTO UIII ACT 10
# IA BARBERO CON LISTA DE DICCIONARIOS
# ==========================================

print("======================================")
print("💈 IA INTELIGENTE DE BARBEROS 💈")
print("======================================")

# Lista vacía de barberos
barberos = []

# Diccionario vacío
losBarberos = {}

while True:

    print("\n====== MENÚ BARBEROS IA ======")
    print("1. Agregar barbero")
    print("2. Ver barberos")
    print("3. Recomendación IA")
    print("4. Eliminar barbero")
    print("5. Salir")

    opcion = input("Selecciona una opción: ")

    # OPCIÓN 1
    if opcion == "1":

        nombre = input("Nombre del barbero: ")
        especialidad = input("Especialidad: ")
        experiencia = input("Años de experiencia: ")

        losBarberos = {
            "nombre": nombre,
            "especialidad": especialidad,
            "experiencia": experiencia
        }

        barberos.append(losBarberos)

        print("✅ Barbero agregado correctamente")

    # OPCIÓN 2
    elif opcion == "2":

        print("\n📋 LISTA DE BARBEROS")

        if barberos == []:

            print("❌ No hay barberos registrados")

        else:

            i = 0

            while i < len(barberos):

                print(barberos[i])

                i = i + 1

    # OPCIÓN 3
    elif opcion == "3":

        print("\n✂ RECOMENDACIÓN IA")

        estilo = input("¿Qué estilo deseas? (clasico/moderno): ")

        if estilo == "clasico":

            print("✅ IA recomienda barbero clásico")

        elif estilo == "moderno":

            print("✅ IA recomienda barbero Fade experto")

        else:

            print("❌ Estilo no válido")

    # OPCIÓN 4
    elif opcion == "4":

        eliminar = input("Nombre del barbero a eliminar: ")

        encontrado = False

        i = 0

        while i < len(barberos):

            if barberos[i]["nombre"] == eliminar:

                del barberos[i]

                encontrado = True

                print("✅ Barbero eliminado")

                break

            i = i + 1

        if encontrado == False:

            print("❌ Barbero no encontrado")

    # OPCIÓN 5
    elif opcion == "5":

        print("\n👋 Gracias por usar la IA de Barberos")
        break

    # ERROR
    else:

        print("❌ Opción inválida")
        print("Programa desarrollado por Paola Silveyra NL = 55")
```
