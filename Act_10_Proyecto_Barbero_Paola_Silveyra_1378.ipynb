#### Usando LISTAS + if + elif + else + while True

```python
# ==========================================
# PROYECTO UIII ACT 10
# IA BARBERO CON LISTA DE DICCIONARIOS
# ==========================================

print("======================================")
print("💈 IA INTELIGENTE DE BARBERO 💈")
print("======================================")

# Lista vacía
clientes = []

# Diccionario vacío
losBarberos = {}

while True:

    print("\n====== MENÚ BARBERO IA ======")
    print("1. Agregar cliente")
    print("2. Ver clientes")
    print("3. Recomendación IA")
    print("4. Eliminar cliente")
    print("5. Ver datos del barbero")
    print("6. Agregar datos del barbero")
    print("7. Salir")

    opcion = input("Selecciona una opción: ")

    # OPCIÓN 1
    if opcion == "1":

        nombre = input("Nombre del cliente: ")
        edad = input("Edad: ")
        cabello = input("Tipo de cabello (liso, rizado, ondulado): ")

        cliente = {
            "nombre": nombre,
            "edad": edad,
            "cabello": cabello
        }

        clientes.append(cliente)

        print("✅ Cliente agregado correctamente")

    # OPCIÓN 2
    elif opcion == "2":

        if clientes == []:

            print("❌ No hay clientes registrados")

        else:

            print("\n📋 LISTA DE CLIENTES")

            i = 0

            while i < len(clientes):

                print(clientes[i])

                i = i + 1

    # OPCIÓN 3
    elif opcion == "3":

        if clientes == []:

            print("❌ No hay clientes para recomendar")

        else:

            cabello = input("Tipo de cabello del cliente: ")

            if cabello == "liso":

                print("✅ IA recomienda Corte Clásico")

            elif cabello == "rizado":

                print("✅ IA recomienda Corte Fade")

            elif cabello == "ondulado":

                print("✅ IA recomienda Corte Texturizado")

            else:

                print("❌ Tipo de cabello no válido")

    # OPCIÓN 4
    elif opcion == "4":

        nombre = input("Nombre del cliente a eliminar: ")

        encontrado = False

        i = 0

        while i < len(clientes):

            if clientes[i]["nombre"] == nombre:

                del clientes[i]

                print("✅ Cliente eliminado")

                encontrado = True

                break

            i = i + 1

        if encontrado == False:

            print("❌ Cliente no encontrado")

    # OPCIÓN 5
    elif opcion == "5":

        if losBarberos == {}:

            print("❌ No hay datos del barbero")

        else:

            print("\n💈 DATOS DEL BARBERO")

            print("Nombre:", losBarberos["barbero"])
            print("Especialidad:", losBarberos["especialidad"])

    # OPCIÓN 6
    elif opcion == "6":

        nombreBarbero = input("Nombre del barbero: ")
        especialidad = input("Especialidad: ")

        losBarberos = {
            "barbero": nombreBarbero,
            "especialidad": especialidad
        }

        print("✅ Datos del barbero guardados")

    # OPCIÓN 7
    elif opcion == "7":

        print("\n👋 Gracias por usar la IA de Barbero")
        break

    # ERROR
    else:

        print("❌ Opción inválida")
        print("Programa desarrollado por Paola Silveyra NL = 55")
```
